## Python Plot Example

Below is a code cell with folding enabled. The code can be hidden, but the plot will be shown.


```python
#| code-fold: true
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.cos(x)

plt.figure(figsize=(8,5))
plt.plot(x, y, color="green", label="Cosine wave")
plt.title("Cosine Function Example")
plt.legend()
plt.show()
```
